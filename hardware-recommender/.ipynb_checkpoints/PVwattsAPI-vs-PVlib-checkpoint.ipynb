## PV Watts API

```python
#shading = .86
#equipDerating = .99
losses = .86
inverterEfficiency = .9

#borough = 'brooklyn'
```

```python
'''
parameters:
latitude,longitude,tilt, azimuth, losses
'''
def pvWattsRequest(lati, longi,t,az,l,nC):
        
    # api documentation: https://developer.nrel.gov/docs/solar/pvwatts/v8/#request-url
    nAzimuth = str(az) #cardinal direction
    nCapkW = str(nC) #nameplate capacity minimum 0.05*
    nLat = str(lati) #latitude
    nLong = str(longi)
    nTilt = str(t)
    nLosses = str(l) #derating*
    nInvEff = str(max(inverterEfficiency,0.9) * 100) #inverter efficiency - cannot be under 90% for API*
    nTimeframe = 'hourly'
    getURL = 'https://developer.nrel.gov/api/pvwatts/v8.json?api_key=' + NREL_API_KEY + '&azimuth=' + nAzimuth+'&system_capacity=' + nCapkW +'&losses=' + nLosses +'&array_type=0&module_type=0&inv_eff=' + nInvEff + '&tilt=' +  nTilt +'&lat=' + nLat +'&lon=' + nLong +'&timeframe=' + nTimeframe

    #print(getURL)
    return getURL

#we care about hourly plane of array irradiance (POA)
# we can also get hourly DC output, but we'll quickly exceed the 1000 calls an hour max...
# *doesn't impact POA

nrelDataRaw = requests.get(pvWattsRequest(lat, long, tilt,azimuth,losses,nameplateCapkW)).json()

nrelData = nrelDataRaw['outputs']

```
